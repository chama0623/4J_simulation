```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("kadai4-4.csv",header=None)
df = df.rename(columns={0:"step",1:"y1",2:"y2"})
df.head()
```

```python
import statsmodels.api as sm

res = sm.tsa.seasonal_decompose(df["y1"],period=15)
trend = res.trend # トレンドデータ
seasonality = res.seasonal # 季節性データ
residual = res.resid # 残差データ

fig = plt.figure(figsize=(20,10),facecolor="white")
fig.suptitle("テスト",fontsize=20)
plt.subplot(4,1,1)
plt.plot(df["step"],df["y1"])
plt.ylabel('Original',fontsize=17)
plt.subplot(4,1,2) 
plt.plot(trend)
plt.ylabel('trend',fontsize=17)
plt.subplot(4,1,3)
plt.plot(seasonality)
plt.ylabel('seasonality',fontsize=17)
plt.subplot(4,1,4) 
plt.plot(residual)
plt.ylabel('residual',fontsize=17)

plt.show()
```
