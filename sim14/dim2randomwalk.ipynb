```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("r2.csv",header=None)
df = df.rename(columns={0:"step",1:"X",2:"Y"})
df.head()
```

```python
plt.figure(facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(df["X"],df["Y"])
```

```python
plt.figure(facecolor="white",dpi=100)
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
lastx = []
lasty = []
for i in range(200):
    df = pd.read_csv("./r2-100/r2-100-"+str(i+1)+".csv",header=None)
    df = df.rename(columns={0:"step",1:"X",2:"Y"})
    # df.head()
    plt.plot(df["X"],df["Y"],color="black")
    lastx.append(df[df["step"]==100]["X"])
    lasty.append(df[df["step"]==100]["Y"])
    
plt.xlabel("X",fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.savefig("r2-100.png",bbox_inches='tight')
```

```python
dist = []
for i in range(200):
    dist.append(lastx[i]**2 + lasty[i]**2)

df100 = pd.DataFrame(dist)
```

```python
df100.describe()
```
