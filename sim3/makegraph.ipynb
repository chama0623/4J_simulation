```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
h1 = pd.read_csv("h1.csv",header=None)
h5 = pd.read_csv("h5.csv",header=None)
h05 = pd.read_csv("h05.csv",header=None)
h1 = h1.rename(columns={0: "t", 1: "x",2:"y"})
h5 = h5.rename(columns={0: "t", 1: "x",2:"y"})
h05 = h05.rename(columns={0: "t", 1: "x",2:"y"})
```

```python
truth_x = np.linspace(-0.5,1.5,1000)
truth_y = np.e**(truth_x)
plt.figure(facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(truth_x,truth_y,label="u = e^t",linewidth = 3.0)
plt.scatter(h1["x"],h1["y"],label="h= 0.1",s=100)
#plt.plot(h1["x"],h1["y"])
plt.scatter(h5["x"],h5["y"],label="h= 0.5",marker="s",s=100)
#plt.plot(h5["x"],h5["y"])
plt.scatter(h05["x"],h05["y"],label="h= 0.05",marker="^",s=100)
#plt.plot(h05["x"],h05["y"])
plt.xlabel("t",fontsize=13)
plt.ylabel("u",fontsize=13)
plt.xlim(-0.1,1.1)
plt.ylim(0.8,3)
plt.legend()
plt.savefig("kadai3.png")
```
