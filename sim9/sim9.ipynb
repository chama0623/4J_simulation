```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter('ignore')
```

```python
df1 = pd.read_csv("h01.csv",header=None)
df2 = pd.read_csv("h001.csv",header=None)
df3 = pd.read_csv("h0001.csv",header=None)
df4 = pd.read_csv("h005.csv",header=None)
df5 = pd.read_csv("h0005.csv",header=None)
df1 = df1.rename(columns={0:"step",1:"X",2:"Y",3:"Vx",4:"Vy",5:"V"})
df2 = df2.rename(columns={0:"step",1:"X",2:"Y",3:"Vx",4:"Vy",5:"V"})
df3 = df3.rename(columns={0:"step",1:"X",2:"Y",3:"Vx",4:"Vy",5:"V"})
df4 = df4.rename(columns={0:"step",1:"X",2:"Y",3:"Vx",4:"Vy",5:"V"})
df5 = df5.rename(columns={0:"step",1:"X",2:"Y",3:"Vx",4:"Vy",5:"V"})
```

```python
df3.tail()
```

```python
df4.tail()
```

```python
plt.figure(figsize=(6,4),facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(df3["X"],df3["Y"],linewidth=1)
plt.grid(False);
plt.xlabel("X",fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.savefig("xy.png",bbox_inches='tight')
```

```python
plt.figure(figsize=(6,4),facecolor="white")
plt.tick_params(direction='in',#目盛り内側
                pad =10, #目盛りとラベルの距離
                length=7.5, #目盛りの長さ
                width=1.5, #目盛りの幅
                labelsize=12,
               )
plt.plot(df5["Vx"],df5["Vy"],linewidth=1)
plt.grid(False);
plt.xlabel("Vx",fontsize=15)
plt.ylabel("Vy",fontsize=15)
plt.savefig("vxy.png",bbox_inches='tight')
```
