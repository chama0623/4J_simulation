```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

## read data

```python
df1 = pd.read_csv("result1.csv",header=None)
df1 = df1.rename(columns={0:"step",1:"X",2:"Y",3:"Z",4:"Vx",5:"Vy",6:"Vz"})
df2 = pd.read_csv("result2.csv",header=None)
df2 = df2.rename(columns={0:"step",1:"X",2:"Y",3:"Z",4:"Vx",5:"Vy",6:"Vz"})
```

## plot
### 3D plot

```python
# 3D
fig = plt.figure(facecolor="white",dpi=100)
ax = Axes3D(fig)
ax.plot(df1["X"], df1["Y"], df1["Z"], "o-", color="#0000aa", ms=2, mew=0.5,clip_on=False)

# 
ax.set_xlabel('x',fontsize=15)
ax.set_ylabel('y',fontsize=15)
ax.set_zlabel('z',fontsize=15)
#plt.savefig("xyz.png",bbox_inches='tight')
# 
plt.show()
```

### plot of each component

```python
plt.figure (facecolor="white",figsize=(12,6),dpi=140)
plt.subplot(3,1,1)
plt.plot(df1["step"],df1["X"])
plt.ylabel("x",fontsize=20)
plt.subplot(3,1,2)
plt.plot(df1["step"],df1["Y"])
plt.ylabel("y",fontsize=20)
plt.subplot(3,1,3)
plt.plot(df1["step"],df1["Z"])
plt.xlabel("$t$",fontsize=15)
plt.ylabel("z",fontsize=20)
plt.savefig("sepxyz.png",bbox_inches='tight')
```

## Confirmation that motion in the z direction is a quadratic function

```python
m = 1 
q = 1
Ez = 1
tscale = np.linspace(0,10.101)
zscale = (q*Ez)/(2*m)* tscale**2 

plt.figure (facecolor="white",dpi=100)
plt.plot(df1["step"],df1["Z"],label="")
plt.plot(tscale,zscale,linestyle="dashed",label="")
plt.xlabel("$t$",fontsize=15)
plt.ylabel("z",fontsize=20)
plt.legend()
plt.savefig("zth.png",bbox_inches='tight')
```
